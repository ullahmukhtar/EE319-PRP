- A sequential experiment involves repeatedly drawing a ball from one of two urns, noting the number on the ball, and replacing the ball in its urn. Urn $0$ contains a ball with the number $1$ and two balls with the number $0$, and urn $1$ contains five balls with the number $1$ and one ball with the number $0$. The urn from which the first draw is made is selected at random by flipping a fair coin. Urn $0$ is used if the outcome is heads and urn $1$ if the outcome is tails. ### Uncountable Sets Revisited
#### *How to generate an event space for the real line?*

The so-called Borel field over the real line, denoted $\mathcal{B}\left(\mathbb{R}\right)$ can be generated by events of the form $\mathsf{B}_{x}=\left(-\infty,x\right]$ that include all subsets we encounter
in practice. Mathematically we write
$$\mathcal{B}\left(\mathbb{R}\right)=\sigma\left(\left\{ \mathsf{B}_{x}:x\in\mathbb{R}\right\} \right)$$

Following are a few examples of events thus generated.
>\begin{align*}
\left(a,\infty\right) & =\overline{\left(-\infty,a\right]}=\overline{\mathsf{B}_{a}}\\
\left(a,b\right] & =\left(a,\infty\right)\cap\left(-\infty,b\right]=\overline{\mathsf{B}_{a}}\cap \mathsf{B}_{b}\\
\left\{ a\right\}  & =\bigcap_{n=1}^{\infty}\left(a-1/n,a\right]=\bigcap_{n=1}^{\infty}\left(\overline{\mathsf{B}_{a-1/n}}\cap B_{a}\right)\\
\left(a,b\right) & =\bigcup_{n=1}^{\infty}\left(a,b-1/n\right]=\bigcup_{n=1}^{\infty}\left(\overline{\mathsf{B}_{a}}\cap B_{b-1/n}\right)\\
\left[a,b\right] & =\bigcap_{n=1}^{\infty}\left(a-1/n,b\right]=\bigcap_{n=1}^{\infty}\left(\overline{\mathsf{B}_{a-1/n}}\cap \mathsf{B}_{b}\right)
\end{align*}

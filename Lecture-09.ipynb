### Dependencies among Component Experiments

- In how many ways can a generous person give five differently-colored M&M's to five kids?

>The important thing to note here is the restriction that each kid can only get one color. If the person gives a red M&M to the first kid, the option no longer exists of giving a red to any of the others. Let us map parts of the problem statement to parts of a placement problem and those of a selection problem.
>

| Problem statement | $\leftrightarrow$ | Placement problem | $\leftrightarrow$ | Selection problem |
| ---- |----|----|----|----|
| "give $\ldots$ differently-colored $\ldots$"   | $\leftrightarrow$ | "place with exclusion" |$\leftrightarrow$ | "sample without replacement" |
| "five kids" |$\leftrightarrow$| "5 distinguishable objects" |$\leftrightarrow$| "ordered selection of 5 elements" |
| "five $\ldots$ M&M's" |$\leftrightarrow$| "5 different cells" |$\leftrightarrow$| "5-element sample space" |

>If you are still not convinced why the M&M's, and not the kids, are our cells, think about where the restriction comes from. Does it come from the kids? Of course not. The restriction comes from the availability of only one color for each kid.

> The multiplicative counting principle applies to the selection problem with dependecy. Thus there are 5 choices (colors) for the first element (kid), 4 for the second, all the way to 1 for the fifth. That gives a total of 
$5\times 4\times 3\times 2\times 1$ ways. These are in fact $5$-permutations.

- How many five-digit numbers ending with 1,2, or 4 are there with no digit repeated?

>

| Problem statement | $\leftrightarrow$ | Placement problem | $\leftrightarrow$ | Selection problem |
| ---- |----|----|----|----|
| "$\ldots$ with no digit repeated"   | $\leftrightarrow$ | "place with exclusion" |$\leftrightarrow$ | "sample without replacement" |

- In how many ways can $n$ people be seated in a row of $n$ chairs?

>Let us map parts of the problem statement to parts of a placement problem and those of a selection problem.
>

| Problem statement | $\leftrightarrow$ | Placement problem | $\leftrightarrow$ | Selection problem |
| ---- |----|----|----|----|
| "seated"   | $\leftrightarrow$ | "place with exclusion" |$\leftrightarrow$ | "sample without replacement" |
| "$n$ people" |$\leftrightarrow$| "$n$ different cells" |$\leftrightarrow$| "$n$-element sample space" |
| "in a row of $n$ chairs" |$\leftrightarrow$| "$n$ distinguishable objects" |$\leftrightarrow$| "ordered selection of $n$ elements" |

> Wonder why "seated" was translated to "placement with exclusion". A person were can be seated on only one chair. No such restriction would apply if, for example,  the person were standing beside or touch a chair. Similary, a row of chairs implies distinguishabilty of chairs. Randomly scattered chairs on a floor with the possibility of some chairs on top of thers would be indistinguishable.   
>In the language of selection problem, there are<br/>
$n$ choices (persons) for chair $1$,<br/>
$n-1$ choices for chair $2$,<br/>
$n-2$ choices for chair $3$,<br/>
$\vdots$<br/>
$n-k+1$ choices for chair $k$,<br/>
$\vdots$<br/>
$1$ choice left for the last chair<br/>
The number of seatings is therefore
$$n!=n\times(n-1)\times(n-2)\cdots 3\times 2\times 1$$
>

- In how many ways can a row of $k$ seats be filled from a set of $n$ people?

$$(n)_k=n(n-1)(n-2)\cdots (n-k+1)$$

 - In how many ways can $n$ people be seated in a circle of $n$ chairs?
 
>Treating this problem as if the chairs were in a row of $n$ chairs, there are $n!$ ways to seat $n$ people. This, however, overcounts the number of ways they can be seated in a circle as each distinct arrangement is counted $n$ times, once for each rotation of the chairs. The problem is said to feature rotational symmetry. To account for this, we must divide the number of arrangements by $n$, yielding $n!/n=(n-1)!$ distinct arrangements.

 - In how many ways can 6 different keys be arranged on a keychain?
 
>This problem also features rotational symmetry and your first guess would be $(6-1)!$ arrangements. This, however, still overcounts the actual arrangements as each distinct arrangement is counted twice, once for each flip of that arrangement. The problem is said to feature reflective symmetry. To account for this, we must divide the number of arrangements by $2$, yielding $(6-1)!/2$ distinct arrangements.

 - In how many ways can a generous person give one M&M each to two children if she has 3 different red, 4 different brown, and 5 different tan ones and the two children insist upon having M&M's of different colors?
 
> Think of a sequential experiment with two component experiments each of the form
$$\Omega \leftrightarrow \text{"give one M&M to one child"}$$
so that the outcome space of the sequential experiment is $\Omega^2$.<br/>
Following are the events of interest for this problem.
\begin{align*}
R & \leftrightarrow\text{"give a red M&M to one child"}\\
B & \leftrightarrow\text{"give a brown M&M to one child"}\\
T & \leftrightarrow\text{"give a tan M&M to one child"}\\
R\times B & \leftrightarrow\text{"give a red-brown pair of M&M's to the children"}\\
B\times R & \leftrightarrow\text{"give a brown-red pair of M&M's to the children"}\\
B\times T & \leftrightarrow\text{"give a brown-tan pair of M&M's to the children"}\\
T\times B & \leftrightarrow\text{"give a tan-brown pair of M&M's to the children"}\\
T\times R & \leftrightarrow\text{"give a tan-red pair of M&M's to the children"}\\
R\times T & \leftrightarrow\text{"give a red-tan pair of M&M's to the children"}
\end{align*}
The required number of arrangements is obtained by summing the cardinalities of the six Cartesian products above, namely
$$\left|R\times B\right|+\left|B\times R\right|+\left|B\times T\right|+\left|T\times B\right|+\left|T\times R\right|+\left|R\times T\right|=2\left(3\times4+4\times5+5\times3\right)=94$$

### Distinguishability Matters

- In how many ways can four objects be put into two boxes?
  - Distinguishable boxes
    - distinguishable objects
    
    >We need to view this as a selection problem if we are to apply the multiplicative counting principle. The "4 distinguishable objects" translates to "ordered selection of 4 elements" and the "two distinguishable boxes" translates to "2-element sample space". Since no sampling/placement restriction been specified, the sampling must be with replacement. There are $2$ choices for each of the $4$ elements, yielding $2^4$ arrangements.

    - indistinguishable 
    
    >Since the objects are indinguishable, all we need to know for each outcome is the number of objects in each box. The only possible outcomes are then $(4,0)$, $(3,1)$, $(2,2)$, $(1,3)$, and $(0,4)$. That amounts to a total of $5$ arrangements.

  - Indistinguishable boxes
    - distinguishable objects 
    
    >Since the boxes are indinguishable, we only need to consider how one box can be filled regardless of without caring about the other box. The box can be empty in only one way, it can hold one object (of the four distinguishable) in four ways, and it can hold two objects in three ways as any specific object (of the four) can be paired with any of the remaining three. That amounts to a total of $1+4+3$ arrangements. If you still do not get it, look at the table following the fourth case.

    - indistinguishable 
    
    >Since the boxes as well as the objects are indinguishable, we only need to consider how the four objects can be split into two groups and list the number of objects in each group. The only possibe groupings are $\{4,0\}$, $\{3,1\}$, and $\{2,2\}$. That amounts to a total of $3$ arrangements.

### Counting Something Else

- The $54$-member Council for Security and Cooperation in Europe wishes to choose $3$ member states for different leadership positions. The $\mathsf{L}$ lobby decrees that at least one of Lithuania, Lichtenstein, Latvia, and Luxembourg must be chosen. In how many ways can the committee be selected?

>The number of arrangements excluding the $\mathsf{L}$ lobby is $(50)_3$. This number, measuring the opposite of what is of interest here, must be subtracted from the total number of arrangements including all the member states, that is, $(54)_3$. The difference $(54)_3-(50)_3$ yields the required number of arrangements.

- [*M&M's Revisited*] In how many ways can a generous person give one M&M each to two children if she has 3 different red, 4 different brown, and 5 different tan ones and the two children insist upon having M&M's of different colors?

> Think of a sequential experiment with two component experiments each of the form
$$\Omega \leftrightarrow \text{"give one M&M to one child"}$$
so that the outcome space of the combined experiment is $\Omega^2$.<br/>
The union of the following events is complementary to what is of interest for this problem.
\begin{align*}
R^2 & \leftrightarrow\text{"give one red M&M each the children"}\\
B^2 & \leftrightarrow\text{"give one brown M&M each to the children"}\\
T^2 & \leftrightarrow\text{"give one tan M&M each to the children"}\\
\end{align*}
The required number of arrangements is obtained by subtracting the cardinalities of the above three events from that of $\Omega^2$, namely
$$\left|\Omega^2\right|-\left|R^2\right|-\left|B^2\right|-\left|T^2\right|=(3+4+5)^2-3^2-4^2-5^2=94$$

### Strategic Overcounting

- In how many ways can Faris choose 3 of the 5 colors of M&M's to eat?

>The problem indicates an unordered selection of 3 elements (colors) from a 5-element sample space. Treating the problem as if the selections were ordered, there are $(5)_3$ ways to choose the colors. This number of $3$-permutations, however, overcounts the number of unordered selections as each distinct selection is counted $3!$ times, once for each rearrangement of the colors. To account for this, we must divide the number of ordered selections by $3!$, yielding $(5)_3/3!$ distinct unordered selections. These are, in fact, $3$-combinations.

-  In how many ways can $k$ objects be selected as a lot a set of n?

>The number of $k$-permutations overcounts the number of $k$-combinations as each combination is counted $k!$ times, once for each rearrangement of the objects in the combination. To account for this, we must divide the number of $k$-permutations by $k!$, yielding $(n)_k/k!$ distinct combinations. The number of $k$-combinations is compactly denoted by the binomial coefficient $\binom{n}{k}$.

-  Find the number of different $k$-element subsets of an $n$-element set. 

>$$\binom{n}{k} = \frac{(n)_k}{k!} = \frac{n!}{k!(n-k)!}=\binom{n}{n-k}$$

- In how many ways can a three-person subcommittee be chosen from a five-person committee?

>$$\binom{5}{3} = \binom{5}{2}$$

- A set of points is chosen on the circumference of a circle so that the number of different triangles with all three vertices among the points is equal to the number of pentagons with all five vertices in the set. How many points are there?

>$$\binom{n}{3} = \binom{n}{5} = \binom{n}{n-5}\Rightarrow n-5=3 \Rightarrow n=8$$

- How many rearrangements of the word LINGUINI are there?

>The word has some letters repeated. Treating the word as if the letters were all different, there are $8!$ rearrangements. This, however, overcounts the number of permutations with repetition as each distinct permutation is counted $3!$ times, once for each rearrangement of $\mathsf{L}$'s, and $2!$ times, once for each rearrangement of $\mathsf{I}$'s. To account for this, we must divide the number of rearrangements of the word by $3!$ and $2!$, yielding $8!/(3!\times 2!)$ distinct permutations.

- Derive a formula for the number of rearrangements of a word consisting of $k_1$ copies of letter $1$, $k_2$ copies of letter $2$, and so on up to $k_m$ copies of letter $m$, with $n$ total letters.

>This number is given by the so-called multinomial coefficient
$$\binom{n}{k_{1},k_{2},\cdots,k_{m}}=\frac{n!}{k_{1}!k_{2}!\cdots k_{m}!}$$

- In how many different ways can an $n$-element set be partitioned into $m$ subsets of cardinalities $\left\{k_1,k_2,\ldots,k_m\right\}$?

>$$\binom{n}{k_{1},k_{2},\cdots,k_{m}}=\binom{n}{k_{1}}\binom{n-k_{1}}{k_{2}}\binom{n-k_{1}-k_{2}}{k_{3}}\cdots\binom{n-\displaystyle\sum_{i=1}^{m-2}k_{i}}{k_{m-1}}=\frac{n!}{k_{1}!k_{2}!\cdots k_{m}!}$$

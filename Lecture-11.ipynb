### The so-called Four Cases of Selection $\leftrightarrow$ Placement
The following definitions are given to fix the experimental settings.
\begin{align*}
\mathsf{\Omega} & =\left\{ \omega_{i}\mid i\in\mathsf{M}\right\} \leftrightarrow\text{‘Base sample space’} & \mathsf{M} & =\left\{ 1,2,\ldots,m\right\} \leftrightarrow\text{‘Index set for different cells’}\\
\mathsf{N} & =\left\{ 1,2,\ldots,n\right\} \leftrightarrow\text{‘Index set for selected elements’} & \mathsf{S} & =\left\{ s_{i}\mid i\in\mathsf{N}\right\} \leftrightarrow\text{‘Set of distinguishable objects’}\\
P\left(\left\{ \omega_{i}\right\} \right) & =p_{i},\ i\in\mathsf{M}\leftrightarrow\text{‘Base probability measure’} &  & \left\{ s\colon n\right\} \leftrightarrow\text{‘Multiset of indistinguishable objects’}
\end{align*}

How the probabilities are assigned to elementary events by a probability
measure is described by what is called a probability distribution.
A uniform probability distribution assigns equal probabilities to elementary
events of a discrete outcome space. In the present setting, the base
probability distribution is uniform if $p_{i}=1/m$ for all $i\in\mathsf{M}$.

- Ordered Selection with Replacement $\leftrightarrow$ Placement of Distinguishable Objects without Exclusion
>For all selection indices $i\in\mathsf{N}$ of the component experiments,
$$\mathsf{\Omega}_{i}=\mathsf{\Omega},\ P_{i}=P$$
The (combined) sample space
\begin{align*}
\mathsf{\Omega}^{n} & =\Bigl\{\left(x_{i}\right)_{i\in\mathsf{N}}\mid x_{i}\in\mathsf{\Omega}\Bigr\}\leftrightarrow\left\{ \left(\mathsf{S}_{i}\right)_{i\in\mathsf{M}}\mid\mathsf{S}_{i}\subseteq\mathsf{S},\sum_{i=1}^{m}\left|\mathsf{S}_{i}\right|=n\right\} 
\end{align*}
The size of the sample space
$$\left|\mathsf{\Omega}^{n}\right|=m^{n}$$
For every outcome $\left(x_{i}\right)_{i\in\mathsf{N}}\in\mathsf{\Omega}^{n}$, the associated elementary event has probability
$$P_{\mathrm{o\text{-}r}}\left(\left\{ \left(x_{i}\right)_{i\in\mathsf{N}}\right\} \right)=P\left(\prod_{i=1}^{n}\left\{ x_{i}\right\} \right)=\prod_{i=1}^{n}P\left(\left\{ x_{i}\right\} \right)$$
>If the number $n_{i}$ copies of $\omega_{i}$ in an outcome is known,
the above probabilities take a more convenient form
$$\prod_{i=1}^{n}P\left(\left\{ x_{i}\right\} \right)=\prod_{i=1}^{m}P\left(\left\{ \omega_{i}\right\} \right)^{n_{i}}=\prod_{i=1}^{m}p_{i}^{n_{i}}$$
>    - A special case of interest is that of $m=2$, $p_{1}=p$ and $n_{1}=k$
which reduces the above probability expression to $p^{k}\left(1-p\right)^{n-k}$.
         - The experiment is said to have $n$ Bernoulli trials, each with probabiliy $p$ of success
         - The above expression then gives the probability of $k$ successes in $n$ Bernoulli trials
>    - Bernoulli trials can arise even when $m>2$ if we have only one event
$\mathsf{R}$ of interest, as our success, at each stage of selection.
We could then consider a partition $\left\{ \mathsf{R},\mathsf{R^{c}}\right\} $
of the base sample space $\mathsf{\Omega}$ with probabilities $P\left(\mathsf{R}\right)=p$
and $\left|\mathsf{R^{c}}\right|=1-p$. If our ordered selection is
known to have $k$ components from $\mathsf{R}$, the associated sequence
of Bernoulli trials has $k$ successes and $n-k$ failures, and the
above probability expression reduces to $p^{k}\left(1-p\right)^{n-k}$. 
>    - If the base probability distribution is uniform, then the elementary events are equiprobable, each with probability $1/m^{n}$.

- Ordered Selection without Replacement $\leftrightarrow$ Placement of Distinguishable Objects with Exclusion
>The sample space
\begin{align*}
\left(\mathsf{\Omega}\right)_{n} & =\left\{ \left(x_{i}\right)_{i\in\mathsf{N}}\mid x_{i}\in\mathsf{\Omega}\setminus\bigcup_{j=1}^{i-1}\left\{ x_{j}\right\} \right\} \leftrightarrow\left\{ \left(\mathsf{S}_{i}\right)_{i\in\mathsf{M}}\mid\mathsf{S}_{i}\subseteq\mathsf{S},\left|\mathsf{S}_{i}\right|\le1,\sum_{i=1}^{m}\left|\mathsf{S}_{i}\right|=n\right\} 
\end{align*}
The size of the sample space
$$\left|\left(\mathsf{\Omega}\right)_{n}\right|=\left(m\right)_{n}=m\left(m-1\right)\cdot\cdots\left(m-n+1\right)=\left(m-n+1\right)^{(n)}$$
>    - Note the use of the falling/rising factorial indicated by the subscript/superscript.
>
>For every outcome $\left(x_{i}\right)_{i\in\mathsf{N}}\in\left(\mathsf{\Omega}\right)_{n}$, the associated elementary event has probability
\begin{multline*}
P_{\mathrm{o\text{-}\bar{r}}}\left(\left\{ \left(x_{i}\right)_{i\in\mathsf{N}}\right\} \right)\\
\quad=P_{1}\left(\left\{ x_{1}\right\} \right)P_{2\mid1}\left(\left\{ x_{2}\right\} \mid\left\{ x_{1}\right\} \right)P_{3\mid1,2}\left(\left\{ x_{3}\right\} \mid\left\{ \left(x_{1},x_{2}\right)\right\} \right)\cdots P_{n\mid1,2,\cdots,n-1}\left(\left\{ x_{n}\right\} \mid\left\{ \left(x_{1},x_{2},\ldots,x_{n-1}\right)\right\} \right)
\end{multline*}
>    - If the base probability distribution is uniform, the sample space
is reduced by a single element at each stage of selection and, consequently,
the elementary events are equiprobable, each with probability
$$\frac{1}{m}\times\frac{1}{m-1}\times\frac{1}{m-2}\cdots\times\frac{1}{\left(m-n+1\right)}=\frac{1}{\left(m\right)_{n}}$$
>    - Furthermore, consider a partition $\left\{ \mathsf{R},\mathsf{R^{c}}\right\} $of the base sample space $\mathsf{\Omega}$ with sizes $\left|\mathsf{R}\right|=r$ and $\left|\mathsf{R^{c}}\right|=m-r$. If the ordered selection is known to have $k$ elements of $\mathsf{R}$, and, obviously, $n-k$ elements of $\mathsf{R^{c}}$, the general product of probabilities above will have $k$ falling factors contributed by elements of $\mathsf{R}$ and $n-k$ falling factors contributed by elements of $\mathsf{R^{c}}$, yielding the special form
$$\dfrac{\left(r\right)_{k}\left(m-r\right)_{n-k}}{\left(m\right)_{n}}$$

- Unordered Selection without Replacement $\leftrightarrow$ Placement of Indistinguishable Objects with Exclusion
>The sample space
\begin{multline*}
\binom{\mathsf{\Omega}}{n}=\left\{ \left\{ x_{i}\mid i\in\mathsf{N}\right\} \mid x_{i}\in\mathsf{\Omega}\setminus\bigcup_{j=1}^{i-1}\left\{ x_{j}\right\} \right\} \\
\qquad\qquad\leftrightarrow\left\{ \left(\mathsf{S}_{i}\right)_{i\in\mathsf{M}}\mid\mathsf{S}_{i}\subseteq\left\{ s\right\} ,\sum_{i=1}^{m}\left|\mathsf{S}_{i}\right|=n\right\} \leftrightarrow\left\{ \left(n_{i}\right)_{i\in\mathsf{M}}\mid n_{i}\in\left\{ 0,1\right\} ,\sum_{i=1}^{m}n_{i}=n\right\} 
\end{multline*}
The size of the sample space
$$\left|\binom{\mathsf{\Omega}}{n}\right|=\binom{m}{n}=\dfrac{\left(m\right)_{n}}{n!}=\binom{m}{m-n}$$
>
>For every outcome $\left\{ x_{i}\mid i\in\mathsf{N}\right\} \in\binom{\mathsf{\Omega}}{n}$, the associated elementary event has probability
\begin{align*}
P_{\mathrm{\bar{o}\text{-}\bar{r}}}\left(\left\{ \left\{ x_{i}\mid i\in\mathsf{N}\right\} \right\} \right) & =P_{\mathrm{o\text{-}\bar{r}}}\left(\Bigl\{\left(y_{i}\right)_{i\in\mathsf{N}}\mid y_{i}\in\left\{ x_{i}\mid i\in\mathsf{N}\right\} \Bigr\}\right)=n!P_{\mathrm{o\text{-}\bar{r}}}\left(\left\{ \left(x_{i}\right)_{i\in\mathsf{N}}\right\} \right)
\end{align*}
>    - If the base probability distribution is uniform, the set $\binom{\mathsf{\Omega}}{n}$
is essentially a contraction of $\left(\mathsf{\Omega}\right)_{n}$
by a factor $n!$, and the elementary events are, still, equiprobable,
each probability
$$\dfrac{n!}{\left(m\right)_{n}}=\dbinom{m}{n}^{-1}$$
>    - Furthermore, consider a partition $\left\{ \mathsf{R},\mathsf{R^{c}}\right\} $of
the base sample space $\mathsf{\Omega}$ with sizes $\left|\mathsf{R}\right|=r$
and $\left|\mathsf{R^{c}}\right|=m-r$. Say our interest lies in the number of elements
of $\mathsf{R}$ in our selection. The appropriate
event is
>$$\mathsf{A}_{k}\leftrightarrow \text{‘} k\text{ elements of }\mathsf{R}\text{ and }n-k\text{ elements of }\mathsf{R^{c}}\text{ in the }n\text{-element selection’}$$
>There are $\dbinom{r}{k}$
ways to select $k$ elements of $\mathsf{R}$ and, independently,
$\dbinom{m-r}{n-k}$ ways to select $n-k$ elements of $\mathsf{R^{c}}$.
Assuming a uniform base probability distribution, the event of our
interest has the probability: 
$$P_{\mathrm{\bar{o}\text{-}\bar{r}}}\left(\mathsf{A}_{k}\right)=\dfrac{\dbinom{r}{k}\dbinom{m-r}{n-k}}{\dbinom{m}{n}}$$
This is an example of a so-called *hypergeometric distribution*.
>    - Derive the hypergeometric distribution from the special form
$$\dfrac{\left(r\right)_{k}\left(m-r\right)_{n-k}}{\left(m\right)_{n}}$$
for an ordered selection.

- Unordered Selection with Replacement $\leftrightarrow$ Placement of Indistinguishable Objects without Exclusion
>The sample space
\begin{multline*}
\biggl(\!\!\binom{\mathsf{\Omega}}{n}\!\!\biggr)=\left\{ \left\{ \omega_{i}\colon n_{i}\mid i\in\mathsf{M}\right\} \mid\omega_{i}\in\mathsf{\Omega},n_{i}\in\left\{ 0\right\} \cup\mathsf{N},\sum_{i=1}^{m}n_{i}=n\right\} \\
\qquad\qquad\leftrightarrow\left\{ \left(\mathsf{S}_{i}\right)_{i\in\mathsf{M}}\mid\mathsf{S}_{i}\subseteq\left\{ s\colon n\right\} ,\sum_{i=1}^{m}\left|\mathsf{S}_{i}\right|=n\right\} \leftrightarrow\left\{ \left(n_{i}\right)_{i\in\mathsf{M}}\mid n_{i}\in\left\{ 0\right\} \cup\mathsf{N},\sum_{i=1}^{m}n_{i}=n\right\} 
\end{multline*}
The size of the sample space
$$\left|\biggl(\!\!\binom{\mathsf{\Omega}}{n}\!\!\biggr)\right|=\biggl(\!\!\binom{m}{n}\!\!\biggr)=\binom{m-1+n}{n}=\frac{\left(m-1+n\right)_{n}}{n!}=\frac{\left(m\right)^{(n)}}{n!}=\binom{m-1+n}{m-1}$$
>
>For every outcome $\left\{ \omega_{i}\colon n_{i}\mid i\in\mathsf{M}\right\} \in\bigl(\!\!\binom{\mathsf{\Omega}}{n}\!\!\bigr)$, the associated elementary event has probability
\begin{multline*}
P_{\mathrm{\bar{o}\text{-}r}}\left(\left\{ \left\{ \omega_{i}\colon n_{i}\mid i\in\mathsf{M}\right\} \right\} \right)\\
\shoveleft{\begin{split}\qquad & =P_{\mathrm{o\text{-}r}}\left(\Bigl\{\left(x_{j}\right)_{j\in\mathsf{N}}\mid x_{j}\in\left\{ \omega_{i}\colon n_{i}\mid i\in\mathsf{M}\right\} \Bigr\}\right)\\
 & =\binom{n}{n_{1},n_{2},\cdots,n_{m}}P_{\mathrm{o\text{-}r}}\left(\prod_{i=1}^{m}\left\{ \omega_{i}\right\} ^{n_{i}}\right)\\
 & =\binom{n}{n_{1},n_{2},\cdots,n_{m}}\prod_{i=1}^{m}p_{i}^{n_{i}}
\end{split}
}
\end{multline*}
The elementary events in this case are not equiprobable. Instead the
probabilities follow what is called a *multinomial distribution*.
>    - Even if the base probability distribution is uniform, the probabilties
$$\binom{n}{n_{1},n_{2},\cdots,n_{m}}\prod_{i=1}^{m}\left(\frac{1}{m}\right)^{n_{i}}=\dfrac{1}{m^{n}}\binom{n}{n_{1},n_{2},\cdots,n_{m}}$$
are not all equal.
>    -In the above forumulation of multinomial distribution, the $m$ elements
of the base sample space are viewed as different varieties or classes.
Multinomial distribution can arise in more complicated situations
in which the $m$ elements of the base sample space are gouped into
$v$ varieties so as to form a partition $\left\{ \mathsf{R}_{1},\ldots,\mathsf{R}_{v}\right\} $
with probabilities $P\left(\mathsf{R}_{i}\right)=\rho_{i}$. The probability
that the unordered repeated selection (of size $n$) has $k_{i}$
elements coming from $\mathsf{R}_{i}$ is given by 
$$\binom{n}{k_{1},k_{2},\cdots,k_{v}}\prod_{i=1}^{v}\rho_{i}^{k_{i}}$$
>    - In the special cases of $\left\{m=2,\,p_{1}=p,\,n_{1}=k\right\}$, or equivalently $\left\{v=2,\,\rho_{1}=p,\,k_{1}=k\right\}$ the probability
distribution reduces to what is called a *binomial distribution*
$$P_{\mathrm{\bar{o}\text{-}r}}\left(\mathsf{R}_{k}\right)=\binom{n}{k}p^{k}\left(1-p\right)^{n-k}$$
where
$$\mathsf{A}_{k}\leftrightarrow \text{‘} k\text{ successes in }n\text{ Bernoulli trials each probability }p\text{ of success’}$$

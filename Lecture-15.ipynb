### Support of a Discrete Distribution
A slight inconvenience about the two point functions, the discrete PDF and CDF, can be noticed. Despite being nondecreasing over the entire real line, the CDF $F$ is strictly increasing only over $\mathcal{X}^{\#}$. Consequently,
despite being nonnegative on the entire real line, the discrete PDF $f$ is strictly positive only at points in $\mathcal{X}^{\#}$. That is why $\mathcal{X}^{\#}$ is said to be the support of $f$. This motivates the introduction of two more functions defined strictly on the support
$\mathcal{X}^{\#}$. The *strictly positive* discrete density function
$$
f^{\#}\colon\mathcal{X}^{\#}\rightarrow\left(0,1\right]\quad\text{defined by}\quad f^{\#}\left(x_{k}\right)=p_{k}
$$
 and the *strictly increasing* distribution function
$$
F^{\#}\colon\mathcal{X}^{\#}\rightarrow\left(0,1\right]\quad\text{defined by}\quad F^{\#}\left(x_{k}\right)=\sum_{i=1}^{k}p_{i}
$$
Equipped with these two new (strictly defined) functions, the (original)
discrete PDF can be expressed as
$$
f\left(x\right)=f^{\#}\left(x\right)\left[x\in\mathcal{X}^{\#}\right]
$$
and the (original) CDF as
$$
F\left(x\right)=F^{\#}\left(\max\left(\mathcal{X}^{\#}\cap\left(-\infty,x\right]\right)\right)
$$
In terms of the strictly positive density, the probability of a Borel
set $\mathsf{B}$ takes the form 
$$
P\left(\mathsf{B}\right)=\sum_{x\in\mathcal{X}^{\#}}f^{\#}\left(x\right)\left[x\in\mathsf{B}\right]
$$

### Most likely outcome
Let us work out the most likely outcome in two discrete distributions encountered earlier.

> **Bernoulli distribution**
>
>The strictly positive density is
$$
f^{\#}\left(k\right)=p^{k}\left(1-p\right)^{1-k}
$$
with support
$$\mathcal{K}=\left\{ 0,1\right\}$$
>
>The most likely outcome is obviously
$$\left[p>1-p\right]=\left[p>1/2\right]$$
>
> **Binomial distribution**
>
>The strictly positive density is
$$
f^{\#}\left(k\right)=\binom{n}{k}p^{k}\left(1-p\right)^{n-k}
$$
with support
$$\mathcal{K}=\left[0\cdot\cdot n\right]$$
>
>The dependence of the discrete density on $k$ can be revealed by
the ratio
$$
\frac{f^{\#}\left(k\right)}{f^{\#}\left(k-1\right)}=\dfrac{\dbinom{n}{k}p^{k}\left(1-p\right)^{n-k}}{\dbinom{n}{k-1}p^{k-1}\left(1-p\right)^{n-k+1}}=\dfrac{\left(n-k+1\right)}{k}\frac{p}{1-p}
$$
It can be seen that $f^{\#}\left(k\right)$ is increasing in $k$, that is, $f^{\#}\left(k\right)\ge f^{\#}\left(k-1\right)$, until
$$
k\left(1-p\right)\le\left(n-k+1\right)p
$$
or
$$
k\le\left(n+1\right)p
$$
The most likely value of $k$, that is the most likely number of successes
in $n$ trials, satisfies
$$
\left(n+1\right)p-1<k_{\max}\le\left(n+1\right)p
$$
The unique solution to this equation is
$$
k_{\max}=\left\lfloor \left(n+1\right)p\right\rfloor 
$$
The most likely relative frequency of successes in $n$ trials, satisfies
$$
p-\frac{1-p}{n}<\frac{k_{\max}}{n}\le p+\frac{p}{n}
$$
The long-run relative frequency of successes reduces to
$$
\lim_{n\rightarrow\infty}\frac{k_{\max}}{n}=p
$$
This relates the long-run relative frequency of successes to the probability of success a single trial.
>
>In many physical situations, the most likely, or expected number of successes, is approximately a fixed number, often denoted $\lambda$, determined by physical constraints. Examples are the expected number of clicks on a website in a fixed duration, the expected number of active users in a fixed duration, the expected number of defects in a fixed length of a bar, the expected number of animals in a fixed area of a forest, and so forth. Dividing such a fixed interval (duration or region) into an arbitrarily large number $n$ of tiny intervals means that success in each tiny interval is rare, that is, of an arbitrarily small probability $p=\lambda/n$. In the language of limits, we say that as $n\rightarrow\infty$, $p\rightarrow0$ so as to keep $\lambda$ constant.
>

### Continuous Distribution
How a continuous distribution arises in practice will be illustrated by the following two examples. First, we revisit the spinning arrow in a wheel for a finer treatment.
> **Spinning arrow in a wheel (continued)**
A more detailed representation of our events of interest is to reserve the interval $\mathcal{X}=\left[0,n\right]$ and map the segments, in order, to the (sub)intervals $\left(0,1\right],\left(1,2\right],\ldots,\left(n-1,n\right]$. For reasons that will become obvious soon, an even better choice is to reserve the interval $\mathcal{X}=\left[0,1\right]$ and map the segments, in order, to the intervals $\left(0,1/n\right],\left(1/n,2/n\right],\ldots,\left(\left(n-1\right)/n,1\right]$ among which the total probability is distributed uniformly so that each subset has probability $1/n$. The remaining subset $\mathbb{R}\setminus\mathcal{X}$ is assigned probability $0$. The CDF $F\left(x\right)$ assumes a value of zero in the open interval $\left(-\infty,0\right)$, rises gradually from $0$ to $1/n$ in $\left(0,1/n\right]$, from $1/n$ to $2/n$ in $\left(1/n,2/n\right]$, all the way up to a rise from $\left(n-1\right)/n$ to $1$ in $\left(\left(n-1\right)/n,1\right]$ and remains at that value in $\left(1,\infty\right)$. The pattern suggests a uniform increase of the CDF from $0$ to $1$ in the interval
$\mathcal{X}$. This becomes more obvious in the limiting case of infinitely large $n$. As $n$ gets larger and larger, each (sub)interval shrinks to a point and the probability increment in each interval approaches zero, and yet the CDF keeps increasing. That is exactly how continuous changes occur. The CDF takes the form
$$F\left(x\right)=\begin{cases}
0 & x<0\\
x & 0\le x<1\\
1 & x\ge1
\end{cases}$$
which can be expressed more succinctly as
$$F\left(x\right)=\min\left\{ 1,x\right\} \left[x\ge0\right]$$
It is interesting to note that the probability of any interval $\left(\left(k-1\right)/n,k/n\right]$
equals its length $1/n$. In other words, the probability-to-length
ratio of each each interval is unity. In the limit, as $n\rightarrow\infty$,
both the probability and the interval length approach zero but the
probability-to-length ratio remains unity. In other words, the continuous
PDF is unity in the interval $\left[0,1\right]$ and zero elsewhere.
Mathematically, we write
$$f\left(x\right)=\frac{\mathrm{d}F}{\mathrm{d}x^{-}}=\left[0<x\le1\right]$$
Thus, a finer treatment of a discrete uniform distribution gave rise to a continuous uniform distribution.

Next, we revisit the geometric distribution with a new perspective.
> **Batches of trials until the first success**
>
>The strictly positive density of the geometric distribution is
$$
f^{\#}\left(k\right)=\left(1-p\right)^{k-1}p^{k}
$$
with support $\mathbb{N}$.
>
>Partition the outcome space $\mathbb{N}$ into $n$-element subsets $\left[1\cdot\cdot n\right],\left[n+1\cdot\cdot2n\right],\left[2n+1\cdot\cdot3n\right],\ldots$, viewed as $n$-trial batches. Map these batches, in order, to the intervals $\left[1/n,1\right],\left[1+1/n,2\right],\left[2+1/n,3\right],\ldots$ and so on. As $n$ gets larger and larger without bound, these intervals approach, in order, $\left(0,1\right],\left(1,2\right],\left(2,3\right],\ldots, $and so on. The probability of no success in $k$ trials, or $x_{k}=k/n$ batches, reads
$$
\left(1-p\right)^{k}=\left(1-\frac{\lambda}{n}\right)^{nx_{k}}\rightarrow\mathrm{e}^{-\lambda x}\ \text{as}\ n\rightarrow\infty
$$
The probability of the first success within the first $k$ trials, or within the first $x_{k}=k/n$ batches, approaches the CDF of a continuous distribution, namely
$$
F\left(x\right)=\left(1-\mathrm{e}^{-\lambda x}\right)\left[x\ge0\right]
$$
The density is recovered as
$$
f\left(x\right)=\frac{\mathrm{d}F}{\mathrm{d}x^{-}}=\lambda\mathrm{e}^{-\lambda x}\left[x>0\right]
$$
Thus, a batch-wise finer treatment of a (discrete) geometric distribution gave rise to a continuous exponential distribution.

Generalizing the two examples above, imagine a subset $\mathcal{X}$ of the real line and a continuous probability measure $P$ that assigns probabilities according to a CDF $F$ that, itself, can be expressed in terms of a density by
$$F\left(x\right)=P\left(\left(-\infty,x\right]\right)=\intop_{-\infty}^{x}f\left(x^{*}\right)\mathrm{d}x^{*}$$
With a continuous PDF available, we can express probability of any Borel set $\mathsf{B}$ by 
$$
P\left(\mathsf{B}\right)=\intop_{\mathsf{B}}f\left(x\right)\mathrm{d}x=\intop_{-\infty}^{\infty}f\left(x\right)\left[x\in\mathsf{B}\right]\mathrm{d}x
$$
Thus, a complete characterization of any continuous probability measure $P$ defined on Borel sets is achieved by only two objects:
 - An uncountable set $\mathcal{X}$ of real numbers
 - A nonnegative continuous density function $f(x)$ that integrates to unity over $\mathcal{X}$

### Support of a Continuous Distribution
As was observed in discrete distributions, a slight inconvenience about the two point functions, the continuous PDF and CDF, can be noticed again. 
Despite being nondecreasing over the entire real line, the CDF $F$
is strictly increasing only over $\mathcal{X}$. Consequently, despite
being nonnegative on the entire real line, the continuous PDF $f$
is strictly positive only at points in $\mathcal{X}$. In other words,
$\mathcal{X}$ is the support of $f$. This motivates the introduction
of two more functions defined strictly on the support $\mathcal{X}$:
the strictly increasing distribution function
$$
F^{\wedge}\colon\mathcal{X}\rightarrow\left(0,1\right]\quad\text{defined by}\quad F^{\wedge}\left(x\right)=P\left(\mathcal{X}\cap\left(-\infty,x\right]\right)
$$
and the strictly positive continuous density function
$$
f^{\wedge}\colon\mathcal{X}\rightarrow\mathbb{R}_{+}\quad\text{defined by}\quad f^{\wedge}\left(x\right)=\frac{\mathrm{d}F^{\wedge}}{\mathrm{d}x^{-}}
$$
 Equipped with these two new (strictly defined) functions, the (original)
continuous PDF can be expressed as
$$
f\left(x\right)=f^{\wedge}\left(x\right)\left[x\in\mathcal{X}\right]
$$
and the (original) CDF as
$$
F\left(x\right)=F^{\wedge}\left(\max\left(\mathcal{X}\cap\left(-\infty,x\right]\right)\right)
$$
In terms of the strictly positive density, the expression for the probability of a Borel set $\mathsf{B}$ takes
the form 
$$
P\left(\mathsf{B}\right)=\intop_{\mathcal{X}}f^{\wedge}\left(x\right)\left[x\in\mathsf{B}\right]\mathrm{d}x
$$

### Mixed Distribution
The discrete and continuous distributions can be merged into a single
mixed distribution with PDF
$$
f\left(x\right)=f^{\#}\left(x\right)\left[x\in\mathcal{X}^{\#}\right]+f^{\wedge}\left(x\right)\left[x\in\mathcal{X}\right]
$$
and CDF
$$
F\left(x\right)=F^{\#}\left(\max\left(\mathcal{X}^{\#}\cap\left(-\infty,x\right]\right)\right)+F^{\wedge}\left(\max\left(\mathcal{X}\cap\left(-\infty,x\right]\right)\right)
$$
Equipped with a single PDF for a mixed distribution, we can express
probability of any Borel set $\mathsf{B}$ by
$$
P\left(\mathsf{B}\right)=\sum_{x\in\mathcal{X}^{\#}}f^{\#}\left(x\right)\left[x\in\mathsf{B}\right]+\intop_{\mathcal{X}}f^{\wedge}\left(x\right)\left[x\in\mathsf{B}\right]\mathrm{d}x
$$

### Measures of size for (un)countable sets
The Lebesgue measure, denoted $\mu^{\wedge}$, of a set generalizes the concept of length of an interval in 1-D, area of a rectangle in 2-D, volume of a cuboid in 3-D, and so on. The Lebesgue measure is useful for uncountable sets but not so for countable sets as it assigns zero value to singletons. To deal with countable sets, the so-called counting measure, denoted $\mu^{\#}$, of a finite set measures its cardinality.

The two measures of the infinitesimal interval
$$
\mathrm{d}\mathsf{B}_{x}=\left(x-\mathrm{d}x,x\right]=\left[x,x+\mathrm{d}x\right)
$$
are then
$$
\mu^{\#}\left(\mathrm{d}\mathsf{B}_{x}\right)=1,\quad\mu^{\wedge}\left(\mathrm{d}\mathsf{B}_{x}\right)=\mathrm{d}x
$$
In terms of these two measures, the probability of a Borel set $B$ takes the form of a so-called Lebesgue-Stieltjes integral
$$
P\left(\mathsf{B}\right)=\intop_{\mathcal{X}^{\#}}f^{\#}\left(x\right)\left[x\in\mathsf{B}\right]\mu^{\#}\left(\mathrm{d}\mathsf{B}_{x}\right)+\intop_{\mathcal{X}}f^{\wedge}\left(x\right)\left[x\in\mathsf{B}\right]\mu^{\wedge}\left(\mathrm{d}\mathsf{B}_{x}\right)
$$
This justifies the usage of the names discrete probability density (derivative of the probability measure with respect to the counting measure) and continuous probability density (derivative of the probability measure with respect to the Lebesgue measure):
$$
f^{\#}=\frac{\mathrm{d}P}{\mathrm{d}\mu^{\#}},\quad f^{\wedge}=\frac{\mathrm{d}P}{\mathrm{d}\mu^{\wedge}}
$$
